```python
#data source https://covid.ourworldindata.org/data/owid-covid-data.json

import pandas as pd
import json
```

```python
df = pd.read_json("https://covid.ourworldindata.org/data/owid-covid-data.json")
```

```python
data = df.transpose()
```

```python
data
```

```python
case = pd.DataFrame()
cont = pd.DataFrame()

for row in range(len(data)):
    loc = data.iloc[row,1]
    temp_df = pd.DataFrame(data.iloc[row,14])
    temp_df["location"] = loc
    #split continent and country
    if(pd.isna(data.iloc[row,0])):
        cont = cont.append(temp_df, ignore_index=True)    
    else:
        temp_df["continent"] = data.iloc[row,0]
        case = case.append(temp_df, ignore_index=True)    
```

```python
case.to_csv("country_data.csv")
cont.to_csv("continent_data.csv")
data.drop("data",axis=1).to_csv("sum_data.csv")
```
